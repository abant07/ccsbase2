```python
import pandas as pd
import joblib

# Load your saved model
model = joblib.load("ccsbase2.joblib")

# Assuming you have a list of feature names (e.g., from your utils.calculate_descriptors)
# Add the 11 one-hot encoded adduct column names
adducts = [
    "[M+H]+", "[M+Na]+", "[M+K]+", "[M+Li]+", "[M+Rb]+", "[M+Cs]+",
    "[M+NH4]+", "[M]+", "[M]-", "[M+dot]+", "[M-Br]+"
]
ohe_adduct_names = [f"Adduct_{a}" for a in adducts]

# 4 non-fingerprint descriptors and OHE adduct columns
feature_names = ["MolWeight", "AdductMass", "Charge", "LabuteASA"] + ohe_adduct_names + ["MorganCountFP"]

# Extract importances
importances = model.feature_importances_

# Create a readable DataFrame
feature_importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print(feature_importance_df.head(10))
```
