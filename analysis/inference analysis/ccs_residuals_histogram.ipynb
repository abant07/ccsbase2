# CCS residuals histogram

This notebook loads `pretrained/testset_predictions.csv`, computes residuals (`CCS_True - CCS_Pred`), and plots a histogram.

```python
import pandas as pd
import matplotlib.pyplot as plt

csv_path = "../pretrained/testset_predictions.csv"

df = pd.read_csv(csv_path)

# Residuals: true minus predicted
residuals = df["CCS_True"] - df["CCS_Pred"]

plt.figure(figsize=(8, 5))
plt.hist(residuals, bins=50, color="#4C72B0", edgecolor="white")
plt.title("Residuals (CCS_True - CCS_Pred)")
plt.xlabel("Residual")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
